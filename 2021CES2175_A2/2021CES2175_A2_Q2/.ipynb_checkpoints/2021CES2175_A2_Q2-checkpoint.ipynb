```python
import numpy as np
import matplotlib.pyplot as plt
E=210000000 #Youngs modulus
I=5*10**-6 # Moment of Interia
W=7
total_nodes = 4
L=[3,4,2]
u = [0,1,0,1,0,1,0,0]
loe = 2 # load on element
F = np.array([0,(-1 * W*L[loe-1]**2)/12,(W*L[loe-1]**2)/12]).reshape((3,1))
load_applied = np.array([-0.5*W*L[loe-1],(-1 * W*L[loe-1]**2)/12,-0.5*W*L[loe-1],(W*L[loe-1]**2)/12]).reshape((4,1))

stiff_matrix = np.zeros((2*total_nodes,2*total_nodes))

matrix_k={}
for j in range(len(L)):
  temp=np.array([[12,6*L[j],-12, 6*L[j]],[6*L[j], 4*L[j]**2, -6*L[j], 2*L[j]**2],[-12, -6*L[j], 12, -6*L[j]],[6*L[j],2*L[j]**2, -6 *L[j], 4*L[j]**2]])
  matrix_k[j] = np.round((E*I/(L[j]**3)) * temp,0)

for i in range(len(L)):
  stiff_matrix[i+i:i+i+4, i+i:i+i+4] += matrix_k[i]

print("Total Stiffness Matrix:\n",stiff_matrix)

iter = 0
for i in u:
  if i==1:
    iter+=1

m_reduced = []
for i in range(2*total_nodes):
  for j in range(2*total_nodes):
    if u[i]*u[j] == 1:
      m_reduced.append(stiff_matrix[i][j])
m_reduced = np.array(m_reduced).reshape((iter,iter))

m_rotation = np.linalg.solve(m_reduced, F)
a=0
b=0
m_disp=[]
for i in u:
  if i == 1:
    m_disp.append(m_rotation[a][0])
    a+=1
  else:
    m_disp.append(u[b])
  b+=1
m_disp = np.array(m_disp).reshape((2*total_nodes,1))
print("Displacement:\n",m_disp)
force_reactions=np.matmul(stiff_matrix, m_disp)
print("force_reactions:\n",force_reactions)

for i in range(len(L)):
  mat_forces = np.matmul(matrix_k[i],m_disp[i+i:i+i+4])
  if i == loe-1:
    mat_forces = mat_forces - load_applied
  print('SF and BM of element ' + str(i+1) +': \n', mat_forces)
  plt.figure()
  plt.title('SFD of Element '+str(i+1))
  plt.plot([0,L[i]],[mat_forces[0],mat_forces[0]])
  plt.savefig("Element_"+str(i+1)+"_SFD")
  plt.figure()
  plt.title('BMD of Element '+str(i+1))
  plt.plot([0,L[i]],[-1*mat_forces[1],-1*mat_forces[2]])
  plt.savefig("Element_"+str(i+1)+"_BMD")
  plt.show()
```
